```python
def MinPCACurveBump(CovMatrix,UsedVar,BumpedVar,Bumps,NumPCA,Locality) :
  #Алгоритм возвращает массив сдвигов кривой

  #CovMatrix - матрица ковариаций приращений относительно изменяемой кривой
  #UsedVar - массив используемых переменных (тип boolean)
  #BumpedVar - сдвигаемые точки кривой
  #Bumps - значения сдвигов
  #NumPCA - число компонент используемых в алгоритме
  #Locality - ?
  if (NumPCA<BumpedVar[UsedVar].sum()) :
    print('Число компонент должно быть больше числа сдвигаемых переменных')
    return np.zeros(shape=UsedVar.size)
  matr=CovMatrix[UsedVar]
  matr=matr.transpose()[UsedVar]
  b=Bumps[BumpedVar==True]
  BumpedVar=BumpedVar[UsedVar]
  c=np.zeros(shape=NumPCA)
  b=np.concatenate([c,b])
  values,vectors=LA.eig(matr)
  V=vectors[:,:NumPCA]*np.sqrt(values[:NumPCA])
  c=np.arange(BumpedVar.size)+1
  c=c*BumpedVar
  c=c[c!=0]-1
  A=np.zeros(shape=(BumpedVar.sum()+NumPCA,BumpedVar.sum()+NumPCA),dtype=np.float64)
  arr=np.zeros(shape=BumpedVar.sum()+NumPCA,dtype=np.float64)
  for i in range(BumpedVar.sum()+NumPCA) :
    if i<NumPCA :
      for j in range(BumpedVar.sum()+NumPCA) : 
        if j==i :arr[j]=-2
        elif j<NumPCA : arr[j]=0
        else : arr[j]=V[c[j-NumPCA],i]
    else :
      for j in range(BumpedVar.sum()+NumPCA) : 
        if j< NumPCA: arr[j]=V[c[i-NumPCA],j]
        else :arr[j]=0
    A[i]=arr.copy()
  B=np.dot(LA.inv(A),b)
  B=B[:NumPCA]
  ans=np.zeros(shape=UsedVar.size)
  ans[UsedVar]=np.dot(V,B)
  return ans
```
